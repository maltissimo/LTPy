```python
def fwhm( nparrayX, nparrayY):
    halfmax = np.max(nparrayY) / 2  # this finds the halfmax  of the array
    # takes the 'derivative' of signum(halfmax - array[])
    d = np.sign(halfmax - nparrayY[0:-1]) - np.sign(halfmax - nparrayY[1:])
    # find the right and left most indexes:
    left_index = np.where(d > 0)[0]
    right_index = np.where(d <= 0)[-1]
    print(nparrayX[right_index])
    print(nparrayX[left_index])
    FWHM = nparrayX[right_index] - nparrayX[left_index]

    return FWHM
    
def calc_2D_fwhm(nparray2D):
    HOR_vector, VER_vector = splitimage(nparray2D)
    #print(len(HOR_vector))
    #print(len(VER_vector))
    HOR_axis = np.arange(0, (len(HOR_vector) + 1), 1)
    VER_axis = np.arange(0, (len(VER_vector)), 1)

    HOR_FWHM = fwhm(HOR_axis, HOR_vector)
    print(len(VER_axis))
    print(len(VER_vector))
    VER_FWHM = fwhm(VER_axis, VER_vector)
    return (HOR_FWHM, VER_FWHM)
    
    
def splitimage( nparray2D):
    center = get_centroid(nparray2D)

    Y_vector = nparray2D[int(center[0]), :]
    X_vector = nparray2D[:, int(center[1])]
    return (X_vector, Y_vector)
```
